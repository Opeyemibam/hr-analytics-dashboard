```python
# check missing values
print(df.isnull().sum())

# remove duplicates
df = df.drop_duplicates()

# convert attrition column
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# confirm
df.head()
```

```python
attrition_rate = df['Attrition'].mean() * 100

print(f"Attrition Rate: {attrition_rate:.2f}%")
```

```python
import matplotlib.pyplot as plt

dept_attrition = df.groupby('Department')['Attrition'].mean().sort_values()

dept_attrition.plot(kind='bar')

plt.title("Attrition Rate by Department")
plt.ylabel("Attrition Rate")
plt.xlabel("Department")

plt.show()
```

```python
import matplotlib.pyplot as plt

# separate groups
attrition_yes = df[df['Attrition'] == 1]
attrition_no = df[df['Attrition'] == 0]

# plot
plt.figure(figsize=(10,6))

plt.scatter(attrition_no.index,
            attrition_no['MonthlyIncome'],
            alpha=0.5,
            label='Stayed')

plt.scatter(attrition_yes.index,
            attrition_yes['MonthlyIncome'],
            alpha=0.7,
            label='Left')

plt.title("Monthly Income vs Attrition")
plt.xlabel("Employee Index")
plt.ylabel("Monthly Income")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# calculate overtime attrition
overtime_attrition = df.groupby('OverTime')['Attrition'].mean()

# plot
overtime_attrition.plot(kind='bar', figsize=(6,5))

plt.title("Attrition Rate by Overtime")
plt.ylabel("Attrition Rate")
plt.xlabel("Overtime")

plt.show()
```

```python
import matplotlib.pyplot as plt

# calculate attrition by role
role_attrition = (
    df.groupby('JobRole')['Attrition']
    .mean()
    .sort_values()
)

# plot
plt.figure(figsize=(10,6))

role_attrition.plot(kind='barh')

plt.title("Attrition Rate by Job Role")
plt.xlabel("Attrition Rate")
plt.ylabel("Job Role")

plt.show()
```

```python
import matplotlib.pyplot as plt

# employees who left
left_employees = df[df['Attrition'] == 1]

# plot
plt.figure(figsize=(10,6))

plt.hist(left_employees['Age'], bins=15)

plt.title("Age Distribution of Employees Who Left")
plt.xlabel("Age")
plt.ylabel("Number of Employees")

plt.show()
```
